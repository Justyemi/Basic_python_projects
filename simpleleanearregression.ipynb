simple linear regression model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # true line with noise
```

```python
 #Add bias term (column of ones)
X_b = np.c_[np.ones((100, 1)), X]
```

```python
# Hyperparameters
learning_rate = 0.1
n_iterations = 1000
```

```python
# Initialize parameters (theta)
theta = np.random.randn(2, 1)
```

```python
# Gradient descent loop
for _ in range(n_iterations):
    gradient = (2/len(X)) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradient
```

```python
print("Intercept:", round(theta[0][0], 2))
print("Slope:", round(theta[1][0], 2))
```

```python
# Plot data and regression line
plt.scatter(X, y, color="blue", label="Data")
plt.plot(X, X_b.dot(theta), color="red", label="Prediction")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
